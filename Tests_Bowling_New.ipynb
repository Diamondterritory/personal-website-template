```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime
import math
```

```python
df = pd.read_csv(r"bowling_data.csv")
del df["Unnamed: 0"]
```

```python
df.head()
```

```python
df = df[df["Captain"].isnull() == False]
df = df[(df["Overs"] != "DNB")&(df["Overs"] != "TDNB")]
df = df.reset_index()
del df["index"]
```

```python
df.head()
```

```python
df["Overs"] = df["Overs"].astype(float)
df["Maidens"] = df["Maidens"].astype(int)
df["Runs"] = df["Runs"].astype(int)
df["Wickets"] = df["Wickets"].astype(int)
df["Economy"] = df["Economy"].astype(float)
df["Innings"] = df["Innings"].astype(int)
```

```python
df.info()
```

```python
def opposition(string):
    return string[2:]
```

```python
df["Opposition"] = df["Opposition"].apply(opposition)
```

```python
def date(string):
    return datetime.strptime(string, "%d-%b-%y")
```

```python
df["Date"] = df["Start Date"].apply(date)
del df["Start Date"]
```

```python
def balls(num):
    return int(((num*10)%10) + ((int(num)*6)))
```

```python
df["Balls"] = df["Overs"].apply(balls)
```

```python
df.head()
```

```python
df2 = pd.read_csv(r"bowlers_data.csv")
del df2['Unnamed: 0']
```

```python
df2.head()
```

```python
df = pd.merge(df,df2,on="Name")
```

```python
df.head()
```

```python
df = df.drop(columns = ["ID","Bowling Style"])
```

```python
df["Pace/Spin"] = df["Speed"]
del df["Speed"]
```

```python
df.head()
```

```python
df = df[["Name","Pace/Spin","Overs","Balls","Maidens","Runs","Wickets","Economy","Innings","Opposition","Venue","Date","Captain"]]
```

```python
df = df.sort_values(by=['Date','Innings'], ascending=[True,True])
```

```python
df.head()
```

```python
df["Strike Rate"] = df["Balls"]/df["Wickets"]
```

```python
summary1 = df.groupby(by=["Captain","Pace/Spin"])["Balls","Maidens","Runs","Wickets"].sum()
```

```python
summary1 = summary1.reset_index()
```

```python
summary1
```

```python
captains = df["Captain"].unique
```

```python
summary2 = df.groupby(by=["Captain"])["Balls","Maidens","Runs","Wickets"].sum()
summary2 = summary2.reset_index()
summary2
```

```python
def tot_balls(captain):
    return int(summary2[summary2["Captain"]==captain]["Balls"])
def tot_maidens(captain):
    return int(summary2[summary2["Captain"]==captain]["Maidens"])
def tot_runs(captain):
    return int(summary2[summary2["Captain"]==captain]["Runs"])
def tot_wickets(captain):
    return int(summary2[summary2["Captain"]==captain]["Wickets"])
```

```python
summary1["Total Balls"] = summary1["Captain"].apply(tot_balls)
summary1["Total Runs"] = summary1["Captain"].apply(tot_runs)
summary1["Total Maidens"] = summary1["Captain"].apply(tot_maidens)
summary1["Total Wickets"] = summary1["Captain"].apply(tot_wickets)
```

```python
summary1
```

```python
summary1["Percent Balls"] = summary1["Balls"]*100/summary1["Total Balls"]
summary1["Percent Maidens"] = summary1["Maidens"]*100/summary1["Total Maidens"]
summary1["Percent Runs"] = summary1["Runs"]*100/summary1["Total Runs"]
summary1["Percent Wickets"] = summary1["Wickets"]*100/summary1["Total Wickets"]
```

```python
def balls_to_overs(balls):
    return (math.floor(balls/6)+((balls%6)*0.1))
```

```python
summary1["Overs"] = summary1["Balls"].apply(balls_to_overs)
```

```python
summary1["Percent Overs"] = summary1["Percent Balls"]
```

```python
summary1
```

```python
import plotly.graph_objects as go
stats = ["Overs","Maidens","Runs","Wickets"]
titles = ["Total overs bowled by bowlers under different captains",
        "Total maidens bowled by bowlers under different captains",
        "Total runs conceeded by bowlers under different captains",
        "Total wickets taken by bowlers under different captains"]
captains = df["Captain"].unique()
for stat in stats:
    fig = go.Figure(data = [
        go.Bar(name = "Pace", x = captains, y = summary1[summary1["Pace/Spin"]=="Pace"][stat]),
        go.Bar(name = "Spin", x = captains, y = summary1[summary1["Pace/Spin"]=="Spin"][stat])
    ])
    fig.update_layout(
        barmode = "group",
        title = titles[stats.index(stat)],
        xaxis_tickangle = -45,
        yaxis = dict(title = stat)
    )
    fig.show()
```

```python
import plotly.graph_objects as go
stats = ["Percent Overs","Percent Maidens","Percent Runs","Percent Wickets"]
titles = ["Percentage of overs bowled by respective categories under different captains",
        "Percentage of maidens bowled by respective categories under different captains",
        "Percentage of runs conceeded by respective categories under different captains",
        "Percentage of wickets taken by respective categories under different captains"]
captains = df["Captain"].unique()
for stat in stats:
    fig = go.Figure(data = [
        go.Bar(name = "Pace", x = captains, y = summary1[summary1["Pace/Spin"]=="Pace"][stat]),
        go.Bar(name = "Spin", x = captains, y = summary1[summary1["Pace/Spin"]=="Spin"][stat])
    ])
    fig.update_layout(
        barmode = "group",
        title = titles[stats.index(stat)],
        xaxis_tickangle = -45,
        yaxis = dict(title = stat)
    )
    fig.show()
```

```python
stats2 = ["Runs","Maidens","Wickets"]
stat_title = ["runs conceeded","maidens bowled","wickets taken"]
for stat in stats2:
    pct_stat = "Percent " + stat
    fig3 = go.Figure(
        data=[
            go.Bar(
                name="Percent overs bowled by pacers",
                x=captains,
                y=summary1[summary1["Pace/Spin"]=="Pace"]["Percent Overs"],
                offsetgroup=0,
            ),
            go.Bar(
                name="Percent overs bowled by spinner",
                x=captains,
                y=summary1[summary1["Pace/Spin"]=="Spin"]["Percent Overs"],
                offsetgroup=0,
                base=summary1[summary1["Pace/Spin"]=="Pace"]["Percent Overs"]
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by pacers",
                x=captains,
                y=summary1[(summary1["Pace/Spin"]=="Pace")][pct_stat],
                offsetgroup=1,
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by spinners",
                x=captains,
                y=summary1[(summary1["Pace/Spin"]=="Spin")][pct_stat],
                offsetgroup=1,
                base=summary1[(summary1["Pace/Spin"]=="Pace")][pct_stat],
            )
        ],
        layout=go.Layout(
            title="Percentage of overs and percentage of " + stat_title[stats2.index(stat)] + " comparison",
            xaxis_tickangle = -45
        )
    )
    fig3.show()
```

```python
def home_away(string):
    if string in away:
        return "Away"
    else:
        return "Home"
```

```python
df["Home/Away"] = df["Venue"].apply(home_away)
```

```python
summary3 = df.groupby(by=["Captain","Home/Away","Pace/Spin"])["Balls","Maidens","Runs","Wickets"].sum()
summary3
```

```python
summary3 = summary3.reset_index()
```

```python
summary3["Overs"] = summary3["Balls"].apply(balls_to_overs)
```

```python
summary3["Total Balls"] = summary3["Captain"].apply(tot_balls)
summary3["Total Runs"] = summary3["Captain"].apply(tot_runs)
summary3["Total Maidens"] = summary3["Captain"].apply(tot_maidens)
summary3["Total Wickets"] = summary3["Captain"].apply(tot_wickets)
```

```python
summary3["Percent Balls"] = summary3["Balls"]*100/summary3["Total Balls"]
summary3["Percent Maidens"] = summary3["Maidens"]*100/summary3["Total Maidens"]
summary3["Percent Runs"] = summary3["Runs"]*100/summary3["Total Runs"]
summary3["Percent Wickets"] = summary3["Wickets"]*100/summary3["Total Wickets"]
```

```python
summary3["Percent Overs"] = summary3["Percent Balls"]
summary3
```

```python
stats2 = ["Runs","Maidens","Wickets"]
stat_title = ["runs conceeded","maidens bowled","wickets taken"]
captains = summary3["Captain"].unique()
for stat in stats2:
    pct_stat = "Percent " + stat
    fig3 = go.Figure(
        data=[
            go.Bar(
                name="Percent overs bowled by pacers at home",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")]["Percent Overs"],
                offsetgroup=0,
                text = round(summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")]["Percent Overs"],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent overs bowled by spinners at home",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Home")]["Percent Overs"],
                offsetgroup=0,
                base=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")]["Percent Overs"],
                text = round(summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Home")]["Percent Overs"],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent overs bowled by pacers away",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Away")]["Percent Overs"],
                offsetgroup=0,
                base=summary3[summary3["Home/Away"]=="Home"].groupby(by=["Captain"])["Percent Overs"].sum(),
                text = round(summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Away")]["Percent Overs"],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent overs bowled by spinners away",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Away")]["Percent Overs"],
                offsetgroup=0,
                base=summary3[(summary3["Home/Away"]=="Home")|(summary3["Pace/Spin"]=="Pace")].groupby(by=["Captain"])["Percent Overs"].sum(),
                text = round(summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Away")]["Percent Overs"],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by pacers at home",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")][pct_stat],
                offsetgroup=1,
                text = round(summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")][pct_stat],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by spinners at home",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Home")][pct_stat],
                offsetgroup=1,
                base=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Home")][pct_stat],
                text = round(summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Home")][pct_stat],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by pacers away",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Away")][pct_stat],
                offsetgroup=1,
                base=summary3[summary3["Home/Away"]=="Home"].groupby(by=["Captain"])[pct_stat].sum(),
                text = round(summary3[(summary3["Pace/Spin"]=="Pace")&(summary3["Home/Away"]=="Away")][pct_stat],1).tolist(),
                hoverinfo="text"
            ),
            go.Bar(
                name="Percent " + stat_title[stats2.index(stat)] + " by spinners away",
                x=captains,
                y=summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Away")][pct_stat],
                offsetgroup=1,
                base=summary3[(summary3["Home/Away"]=="Home")|(summary3["Pace/Spin"]=="Pace")].groupby(by=["Captain"])[pct_stat].sum(),
                text = round(summary3[(summary3["Pace/Spin"]=="Spin")&(summary3["Home/Away"]=="Away")][pct_stat],1).tolist(),
                hoverinfo="text"
            )
        ],
        layout=go.Layout(
            title="Percentage of overs and percentage of " + stat_title[stats2.index(stat)] + " comparison",
            xaxis_tickangle = -45,
            hovermode = "closest"
        )
    )
    fig3.show()
```
